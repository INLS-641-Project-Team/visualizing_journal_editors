## we see that some of the countries have the same value twice like ['USA', United States'] so we need to clean those values

```python
def clean_countries(country):
    
    country = country.strip()
    country2 = re.sub(r"[\[\]']", "", country)
    country2 = re.sub(r'USA|US', "United States", country2)
    country2 = re.sub(r'UK', 'United Kingdom', country2)
    country2 = re.sub(r'Russian Federation', 'Russia', country2)
    country2 = re.sub(r'Czechia', 'Czech Republic', country2)
    country2 = re.sub(r'Ivory Coast|CÃ\´te dIvoire', "Republic of Côte d'Ivoire", country2)
    country3 = np.unique(np.array(country2.split(r', ')))
    country4 = ', '.join(country3) 
    
    return country4

    
    
df['country'] = df['country'].apply(clean_countries)
    
```
